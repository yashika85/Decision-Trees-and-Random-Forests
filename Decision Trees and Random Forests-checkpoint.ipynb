```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import (DecisionTreeClassifier, DecisionTreeRegressor,
                          plot_tree, export_graphviz)
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (accuracy_score, roc_auc_score,
                             r2_score, mean_squared_error)
import graphviz
```

```python
df = pd.read_csv(r"C:\Users\ABC\Downloads\diabetes.csv")
df
```

```python
X  = df.drop(columns=["disease_progression"])
y  = df["disease_progression"]
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(Xtr, ytr)
r2 = r2_score(yte, dt_reg.predict(Xte))
print(f"Decision Tree  |  RÂ²: {r2:.3f}")
```

```python
depths = range(1, 21)
cv_r2  = [cross_val_score(
             DecisionTreeRegressor(max_depth=d, random_state=42),
             X, y, cv=5, scoring="r2").mean()           
          for d in depths]

plt.plot(depths, cv_r2, marker="o")
plt.xlabel("max_depth"); plt.ylabel("CV RÂ²")
plt.title("Decisionâ€‘Tree Regression â€” depth vs generalisation"); plt.show()
```

```python
rf_reg = RandomForestRegressor(n_estimators=300, random_state=42)
rf_reg.fit(Xtr, ytr)
pred   = rf_reg.predict(Xte)
r2   = r2_score(yte, pred)                                    
rmse = np.sqrt(mean_squared_error(yte, pred))                 

print("Random Forest  |  RÂ²:",      round(r2,   3),
      "|  RMSE:",                   round(rmse, 1))
```

```python
(pd.Series(rf_reg.feature_importances_, index=X.columns)
   .sort_values()
   .plot(kind="barh", title="RF â€“ feature importance (regression)")); plt.show()
```

```python
y_bin = (y > y.median()).astype(int)      
Xtr_c, Xte_c, ytr_c, yte_c = train_test_split(
    X, y_bin, test_size=0.25, stratify=y_bin, random_state=42)
```

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(Xtr_c, ytr_c)
pred_c = dt_clf.predict(Xte_c)
acc = accuracy_score(yte_c, pred_c)
print("Decision Tree | accuracy:", round(acc, 3))
```

```python
plt.figure(figsize=(14,6))
plot_tree(dt_clf, feature_names=X.columns,
          class_names=["low","high"], filled=True, rounded=True)
plt.show()
# or Graphviz (higherâ€‘quality SVG/PDF):
dot = export_graphviz(dt_clf, out_file=None, feature_names=X.columns,
                      class_names=["low","high"], filled=True, rounded=True)
graphviz.Source(dot)
```

```python
depths = range(1, 21)
cv_acc = [cross_val_score(
             DecisionTreeClassifier(max_depth=d, random_state=42),
             X, y_bin, cv=5, scoring="accuracy").mean()
          for d in depths]
plt.plot(depths, cv_acc, marker="o")
plt.xlabel("max_depth"); plt.ylabel("CV accuracy")
plt.title("Decisionâ€‘Tree Classifier â€” depth vs generalisation"); plt.show()
```

```python
rf_clf = RandomForestClassifier(n_estimators=300, random_state=42)
rf_clf.fit(Xtr_c, ytr_c)
pred_rf = rf_clf.predict(Xte_c)
acc  = accuracy_score(yte_c, pred_rf)
auc  = roc_auc_score(yte_c, rf_clf.predict_proba(Xte_c)[:,1])

print("Random Forest | accuracy:", round(acc, 3),
      "|  ROCâ€‘AUC:",               round(auc, 3))
```

```python
(pd.Series(rf_clf.feature_importances_, index=X.columns)
   .sort_values()
   .plot(kind="barh", title="RF â€“ feature importance (classification)")); plt.show()
```

```python
param_grid = {"max_depth":[None, 4, 8, 12],
              "min_samples_leaf":[1, 3, 5]}
grid = GridSearchCV(RandomForestRegressor(n_estimators=300, random_state=42),
                    param_grid, cv=5, n_jobs=-1, scoring="neg_mean_squared_error")
grid.fit(X, y)
print(f"Best CV MSE: {-grid.best_score_:.3f}",    
      "\nBest params:", grid.best_params_)
```
